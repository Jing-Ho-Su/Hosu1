```python
#학점이 90 이상이면 A, 80이상 90 미만이면 B,
#  70 이상 80미만이면 C, 60 이상 70 미만이면 D,
#  60 미만이면 F을 출력하는 프로그램을 만드시오
학점 = int(input("점수를 입력하시오 :"))


if 학점 >= 90:
    점수 = "A"
elif 90 > 학점 >=80:
    점수 = "B"
elif 80 > 학점 >=70:
    점수 = "C"
elif 70 > 학점 >=60:
    점수 = "D"
elif 60 > 학점:
    점수 = "F"

print(점수)
```

```python
score = input("학점을 입력하세요 :")

if score >= 90:
    print(f"입력값은 {score} 당신의 학점은 : A")
elif score >= 80:
    print(f("입력값은 {score} 당신의 학점은 : B"))
elif score >= 70:
    print(f("입력값은 {score} 당신의 학점은 : C"))
elif score >= 60:
    print(f("입력값은 {score} 당신의 학점은 : D"))
else:
    print(f("입력값은 {score} 당신의 학점은 : F"))
```
