```python
# enumerate í•¨ìˆ˜ í™œìš©í•˜ê¸° #ë°˜ë³µ ê°€ëŠ¥í•œ ê°ì²´(iterable)ë¥¼ ë°›ì•„ì„œğŸ‘‰ (ì¸ë±ìŠ¤, ê°’) ìŒì„ ì°¨ë¡€ëŒ€ë¡œ ë§Œë“¤ì–´ì£¼ëŠ” í•¨ìˆ˜ì•¼
fruits = ["apply", "banana", "orange"]

for fruit in fruits:
    print(fruit)

for (i, fruit) in enumerate(fruits):
    print(f"{i}:{fruit}")
```
