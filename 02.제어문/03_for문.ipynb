```python
# For문의 기본구조

# for문은 리스트의 개수 만큼 수행할 문장을 반복한다.


# for 변수 in 리스트(또는 튜플, 문자열):
#     수행할 문장1
#     수행할 문장2

list = [1, 2, 3]
```

```python
# 전형적인 for 문
test_list = ["one", "two", "three"]

for i in test_list:
    print(i)
```

```python
# 다양한 for 문의 사용
a = [(1, 2), (3, 4), (5, 6)]

for a,b in a:
    print(a + b)
```

```python
# for 문의 응용

# 5명의 시험점수
scores = [90, 25, 67, 45, 80]
id = 0

for score in scores:
    id += 1 # id = id + 1
    if score >= 60:
        print(f"id : {id}, 합격입니다.")
    else :
        print(f"id:{id}, 불합격입니다.")
```

```python
# for문과 continue문
scores = [90, 25, 67, 45, 80]
id = 0

for score in scores:
    id += 1
    
    if score < 60:
        continue  # 코드 실행 조건을 만족하지 않으면 이후의 코드를 실행하지 않고 위로 올라가서 for문을 다시 처음부터 실행한다
                  # 즉, continue는 이후의 문장을 실행하지 않고, 이전의 문장으로 돌아가서 다시 처음부터 시작하는 연산자이다.
    print(f"id가 {id}인 학생은 합격입니다.")
```

```python
# for문과 함께 자주 사용하는 range 함수

#range는 지정된 범위 사이의 숫자들을 의미한다.
a = range(10)
print(list(a))


b = range(1, 10)
print(list(b))
```

```python
# 1부터 10까지 더하는 반복문

합 = 0

for i in range(1, 10+1): # [1, 2, 3, ... 10]
    합 = 합 + i
    print(합)  # for 반복문이 반복되는 중 합을 출력
print(합) # 마지막 메모리 상의 전체 합을 출력

```

```python
# for 와 range()를 이용한 구구단을 만들어 보세요.
# 출력 예시는 

# a의 반복문이 먼저 반복되고 b의 반복문이 끝까지 반복 된 후에 a의 두번째 반복문으로 넘어간다.
# 그렇기에 a의 첫번째 변수인 2가 그대로 유지되고 이후에 3으로 넘어가고 쭉쭉 9까지 넘어가는 식으로 진행된다.
# 즉 더 상위에 있는 for문은 아래 하위에 있는 for문이 다 반복 끝나야 다음 반복문으로 넘어가는 것이다.

for a in range(2, 10):     # [2,3,4,...9]
    for b in range(1, 10): # [1.2.3,...9] 
        print(f"{a}*{b}={a*b}")
    
```
