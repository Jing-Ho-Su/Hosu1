```python
# 함수를 지정하고 그 지정된 함수에 수를 넣는 것, 즉 함수란 input > output 에서 > 에 해당하는 것인데 > 에 해당하는 것이 함수이고
# input을 지정해 그 값을 얻어내는 것임 이때 input의 함수는 def 의 함수 문에 적으면 안된다 왜냐하면 같이 함수로 정의 되어버리면
# 인수로 사용할 수 없기 때문이다.


#함수란 무엇인가?
## 믹서기, 자판기

#파이썬 함수의 구조

# def 함수이름(매개변수):
#     수행할 문장1
#     수행할 문장2
#     ...

def add(a, b):
    return a + b

result = add(3, 4)
print(result)


```

```python
# 일반적인 함수
## 입력값 : O, 반환값 : O

# def 함수이름(매개변수):
#     수행할문장1
#     수행할문장2
#     ...
#     return 결과값

# 예제       ## def 함수 밖에서 내가 a, b 의 함수를 정해주고 그 수르 넣어주면 함수에서 결과를 작성해 리턴해준다.
def add(a,b):
    result = a + b
    return result

a = add(3, 4)
print(a)
```

```python
# 입력값이 없는 함수  ## 입력값이 없기 때문에 ()안의 매개변수도 없고 함수 안에 리턴 뒤에 저장된 'Hi'가 그냥 출력되었다.
## 입력값 : X, 반환값 : O 

def say():
    return "Hi"

a = say()
print(a)
```

```python
# 반환값이 없는 함수
## 입력값 : O, 반환값 : X

def add(a, b):
    print(f"{a},{b}의 합은 {a+b}이다")

a = add(3,4)
print(a)
```

```python
 ## 입력값 : X, 반환값 : X

def say():
    print("Hi")

a = say()
print(a)
```

```python
# 매개변수 지정하여 호출하기
def sub(a, b):
    return a - b

result1 = sub(7, 3)
print(result1)

result2 = sub(b=7, a=3)
print(result2)

print(sep=" ", end=" ")
```

```python
# # 입력값이 몇 개가 될 지 모를 때는 어떻게 해야 할까?

# def 함수이름(*args):
#     수행할문장1
#     수행할문장2
#     ...

def add_many(*args): # args = [1, 2, 3, 4, 5]
    result = 0
    for i in args:
        result = result + i
    return result

a = add_many(1, 2, 3, 4, 5) # 위의 함수를 다 끝내고 나면 현재 add_many(1,2,3,4,5)= 15가 들어가게된다
print(a)
```
