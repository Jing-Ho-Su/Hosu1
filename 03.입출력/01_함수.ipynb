```python
# # 입력값이 몇 개가 될 지 모를 때는 어떻게 해야 할까?

# def 함수이름(*args):
#     수행할문장1
#     수행할문장2
#     ...

#args 의 앞에 붙는 *가 여러 개의 이자를 묶어서 하나의 튜플로 묶어주고 이를 args에
#할당한다라고 생각하면 된다.

#현재 코드에서 add_many()의 괄호 안의 값에 어떤 수가 들어갈지 몇개가 들어갈지 모르니
#여기에 *args를 작성해준다 즉 수학에서 변수 n이라고 생각하면 편하다.
# 이렇게 정해준 수식을 통해 내가 나중에 어떤 어떤 수를 넣고 싶다고 결정되면
# 나중에 할당해주면된다 이 할당해주는 과정이 a = add_many(1,2,3,4,5)이라고 생각하자.

def add_many(*args): # args = (1, 2, 3, 4, 5)
    result = 0
    for i in args:
        result = result + i
    return result

a = add_many(1, 2, 3, 4, 5) # 위의 함수를 다 끝내고 나면 현재 add_many(1,2,3,4,5)= 15가 들어가게된다
print(a)
```

```python
# 키워드 매개변수, kwargs
## 키워드값의 형태로 매개변수를 받을 때
## 매개변수 앞에 **(2개)를 붙인다

# *args는 알수없는 "숫자"를 대신하는 매개변수 or 변수 였다면
# **kwargs는 알 수 없는 여러개의 '문자=키워드'에 해당하는 것을 대신하는 매개변수인 것이다.
# 근데 여기서 이 **kwargs의 특징은 할당되어 나오는 모습이 "딕셔너리" 형태로 나오게 된다. 

def print_kwargs(**kwargs):
    print(kwargs)


print_kwargs(a=1) # 딕셔너리 형태로 나옴
print_kwargs(name="foo", age=3)
```

```python
def creat_profile(**profiles):
    print("===== 프로필 정보 =====")
    for key, value in profiles.items():
        print(f"{key} : {value}")

creat_profile(이름="김철수", 나이=30, 직업="개발자", 취미="운동")
```

```python
# *args, *kwargs 혼용


# 처음에 " 홍길동"은 그냥 name 에 할당 되었음
# *args 는 n개의 수를 할당하는 변수이니 숫자만 할당함
# **args는 n개의 키워드를 할당하는 변수이고 딕셔너리를 할당하는 매개변수이니 뒤의 key, value 형태의 요소만 할당


def mixed_function(name, *args, **kwargs):
    print(f"이름 : {name}")
    print(f"추가인수들 : {args}")
    print(f"키워드인수들 : {kwargs}")

mixed_function("홍길동", 1, 2, 3, 4, 5, age=25, city = "서울")
```

```python
# return 여러개가 나올 때 행이 나눠지면
# 뒤에 작성된 return은 실행되지 않는다.

def add_and_mul(a,b):
    return a+b
    return a*b

result = add_and_mul(3, 4)
print(result)
```

```python
# return 뒤에 아무것도 작성되어 있지 않는다 즉 nick == 바보라면 그냥 이 실행을 끝낸다는 의미기이에
# 아무값도 나오지 않게 된다.

def say_nick(nick):
    if nick == "바보":
        return
    print(f"나의 별명은 {nick}입니다.")

# say_nick("야호")
say_nick("바보")
```

```python
# 매개변수에 초기값을 미리 설정하기
def say_myself(name, age, man=True):
    print(f"나의 이름은 {name}입니다.")
    print(f"나이는 {age}살 입니다.")
    if man:
        print("남자입니다.")
    else:
        print("여자입니다.")

say_myself("홍길동", 30)
# man으 True로 설정해놨기에 남자입니다.가 출력된다.
say_myself("영희", 30, False)
```

```python
# # lambda 예약어
# def 함수이름(매개변수):
#     수행할문장
#     return

# def add(a, b):
#     return a + b


# lambda
# 함수이름 = lambda 매개면수1, 매개변수2, ... : 수행할문장

add = lambda a, b: a + b
result = add(3, 4)
print(result)
```

```python
def add(a , b):
    """
    함수의 독스트링(Docstring)
    두 숫자를 더하는 함수
    설명 : 다른 사람 또는 나를 위한 문서()
    """

    return a+b

result = add(3, 4)
print(result)

print(add.__doc__)
```
