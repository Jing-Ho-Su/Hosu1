```python
f = open("data2.txt", "w")
f.close()

# 파일객체 = open(파일이름,파일열기모드)

# r - 읽기모드 : 파일을 읽기만 할 때 read
# w - 쓰기모드 : 파일의 내용을 쓸 때 write
# a - 추가모드 : 파일의 마지막에 내용을 추가할 때 append
```

```python
f = open("./data/새파일.txt","w")

# ./ 는 현재 파일이라는 의미


for i in range(1, 11):
    data = f"{i}번째 줄입니다.\n" #\n은 줄바꿈의 의미
    f.write(data)

f.close()
```

```python
f = open("data/새파일.txt", "r")
while True:
    line = f.readline()
    if not line : break
    print(line)
f.close()
```

```python
# readlines() 함수 이용하기 # 얘는 다 읽는 함수지만 list 형식으로 파일을 읽는 함수이다.

f = open("data/새파일.txt", "r")
lines = f.readlines()

for line in lines:
    print(line, end="")
f.close()





print(type(lines))
print(len(lines))
```

```python
# read 함수 사용하기
f = open("data/새파일.txt", "r")
data = f.read()
print(data)
f.close()
```

```python
# 파일에 새로운 내용 추가하기  #../ 위 폴더로 이동하기
f = open("../새파일.txt", "a")  

for i in range(21,30):
    f.write(f"{i}번째 줄입니다.\n")

f.close()

```

```python
# with 문과 함께 사용하기
with open("data3.새파일.txt", "w") as f:
    f.write("Life is too short, You need python")
```
