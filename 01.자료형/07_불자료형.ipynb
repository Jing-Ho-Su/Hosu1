```python
# 자료형의 참과 거짓

#문자열 중에서 아무것도 없는값은 false 이고 나머지 모든 문자값은 모두 true 이다
print(bool("python"))
print(bool(""))

#논리식이 들어갈 자리에 빈 리스트가 들어가 있으면 모두 false 이고 그 외 모두 true 이다
print(bool([1,2,3]))
print(bool([]))

#
print(bool((1,2,3)))
print(bool(()))

#
print(bool({"a" : 1}))
print(bool({}))
# 즉 비어있는 형태를 빼면 딕셔너리는 모두 트루이다


# 숫자는 0을 제외하고 모두 true 이다. 즉 0만 false 이다.
print(bool(1))
print(bool(0))



# none은 아무것도 없는 값을 의미하기에 false 값이 된다.
print(bool(None))


```

```python
# and 연산자,  or 연산자

print(True and True) # 트루와 트루의 결합 은 트루
print(True and False) 
print(False and True) 
print(False and False)# 하나라도 false 이면 false 이다 


print(True or True) #or은 하나라도 true가 있으면 true가 나온다
print(True or False) 
print(False or True) 
print(False or False) 


```
