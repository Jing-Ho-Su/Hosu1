```python
#dictionary 사전 
#딕셔너리는 어떻게 만들까?

# 쉬프틑 탭을 누르면 요소가 앞에 붙는다
dic = {
    "name" : "json",
    "phone" : "010-1234-1234",
    "birth" : "00-01-01"
}

#{key: value)


```

```python
# 딕셔너리 쌍 추가, 삭제하기
dic["hobby"] = "running"
dic
```

```python
dic[3] = [1,2,3]
dic


# 딕셔너리에 문자값뿐만아니라 리스트 값의 형태과 숫자 값을 둘다 넣을 수 있다
# 딕셔너리의 key 값에는 list 값을 넣을 수 없지만 value 에는 list 형태의 정보를 넣을 수 있다
```

```python
# 딕셔너리 요소 삭제하기


del dic[3]
print(dic)
```

```python
del dic["hobby"]
print(dic)
```

```python
#딕셔너리 요소 변경하기
dic["name"] = "python"
dic
```

```python
# 딕셔너리에서 key를 사용해서 value 얻기 / key를 적어주면 value 값을 얻을 수 인느게 dic의 특징이다
#드래그를 한 상태에서 괄호를 넣으면 전부 괄호된다
#알트 쉬프트 아래 방향기 를 하면 전부 복사된다

# hobby는 현재 dic에서 없기 때문에 keyerror 라고 뜨고 hobby의 dic은 뜨지 않는다
print(dic["name"])
print(dic["birth"])
print(dic["phone"])
print(dic["hobby"])

```

```python
# 딕셔너리와 관련된 함수
## key list 만들기 - keys

#keys는 dic의 key값을 가지고 오는 함수이다



dic = {
    "name" : "json",
    "phone" : "010-1234-1234",
    "birth" : "00-01-01"
}


dic.keys()       #dic의 값 중 key 값만 가지고 옴
list(dic.keys()) #list를 앞에 쓰니깐 완전한 list형태의 데이터로 바뀜
```

```python
# Value 리스트 만들기 - values
## value만 가지고 오는 함수

li = dic.values() # dic의 값 중 value 값만 가지고 오는 함수를 사용
li                # li 값을 출력하면 value 값만 가지고 옴
list(li)          # 이를 list 값으로 출력하기 위해 list 형식으로 변형
```

```python
#key, value 쌍 얻기 -items
li = dic.items()     #items는 key 값과 value 값을 쌍으로 얻는 함수
list(li)             #튜플의 형태로 얻어짐
```

```python
# key : value 쌍 모두 지우기 - clear

a = dic.clear() # 모든 요소를 다 지움
dic 
a               # 모든 요소를 다 지웠기에 아무것도 출력이 안된다
```

```python
# key로 value 얻기 - get

dic = {
    "name" : "json",
    "phone" : "010-1234-1234",
    "birth" : "00-01-01"
}

# 드래그 하고 컨트롤 슬래시를 하면 주석 처리가 된다
# dic["name"]
print(dic.get("name"))
print(dic.get("phone"))
print(dic.get("brith"))

dic
```

```python
# 해당 key가 딕셔너리 안에 있는지 조사하기 - in

dic = {
    "name" : "json",
    "phone" : "010-1234-1234",
    "birth" : "00-01-01"
}

print("name" in dic)
print("phone" in dic)
print("email" in dic)
print("brith" in dic)


```

```python
# key로 value 얻기 - pop

# get과의 차이점 / get은 단지 딕셔너리의 값을 출력하는 것이고 pop은 출력해서 빼내는 것이라고 생각하면 편하다
## pop는 얻은 key가 dic에서 빠져나간다 그래서 마지막에서 dic를 출력해보면 phone의 값이 빠져나가서 출력이된다


dic = {
    "name" : "json",
    "phone" : "010-1234-1234",
    "birth" : "00-01-01"
}

phone = dic.pop('phone')
print(phone)

print(dic)
```
