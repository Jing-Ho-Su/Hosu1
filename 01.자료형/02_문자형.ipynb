```python
#문자열 인덱싱
#컴퓨터는 0부터 시작함 / 마지막 문자는 -1을 하면 마지막 문자를 가지고 옴

a="life is too short"

print(a[0])
print(a[8])
print(a[-1])
print(a[-2])
```

```python
#문자열 슬라이싱
a='''life is too short, you need python'''

print(a[0:4])
print(a[12:17])
print(a[5:])
print(a[:17])
print(a[:])
```

```python
# 문자열 변수 출력 {}를 사용하면 변수다 라는 뜻임 ,  즉 변수의 값이 프린트 되어 나옴

day = "three"
print(f"i was sick for 3 {day} days")
```
