```python
t1 = (1, 2, "a", "b")

# 튜플 요소갑을 삭제하려할 때 > 지울 수 없음 (튜플의 특징)
del t1[0]
```

```python
# 인덱싱 하기
t1 = (1, 2, "a", "b")
t1[2]
```

```python
# 슬라이싱하기
t1 = (1, 2, "a", "b")
t1[2:3+1] # t1[2:] t1[2:4]
```

```python
# 튜플 더하기 / 튜플은 삭제와 변경은 불가능하지만 '더하기'는 가능하다
t1 = (1, 2, "a", "b")
t2 = (3, 4)

t1 + t2
```

```python
# 튜플의 곱하기 / 튜플은 삭제와 변경은 불가능하지만 '곱하기'도 가능하다
t1 = (1, 2, "a", "b")
t2 = (3, 4)
t2 *3


```
