```python
### 문제 3-1. 성적 관리 프로그램
#학생들의 성적을 관리하는 프로그램을 작성하시오.

# 다음 학생들의 성적 정보를 딕셔너리로 저장하시오
# 학생1: 이름=김민수, 국어=85, 영어=90, 수학=78
# 학생2: 이름=이영희, 국어=92, 영어=88, 수학=95
# 학생3: 이름=박철수, 국어=78, 영어=85, 수학=82

# 요구사항:
# 1. 각 학생의 총점과 평균을 계산하여 딕셔너리에 추가
# 2. 각 학생의 정보를 보기 좋게 출력
# 3. 평균이 가장 높은 학생 찾기

student = {
    '학생1' : {'이름' : '김민수', '국어' : 85, '영어' : 90, '수학' : 78}, 
    '학생2' : {"이름" : "이영희", "국어" : 92, "영어" : 88, "수학" : 95}, 
    '학생3' : {"이름" : "박철수", "국어" : 78, "영어" : 85, "수학" : 82}
}

#for은 여러 개 있는 것을 하나씩 꺼내서 반복 실행하라는 뜻이며
# for x in student.values()는 student의 value 값을 하나씩 꺼내서
# x의 변수라는 이름으로 반복해서 사용하라는 뜻
for x in student.values() :
    total = x["국어"] + x["영어"]+ x["수학"]
    avg = total / 3
    x["총점"] = total
    x["평균"] = total / 3


print("===학생 성적 관리===")
print(student)
print(max(student))



### 문제 3-2. 단어 카운터
#문자열에서 각 단어가 몇 번 등장하는지 세는 프로그램을 작성하시오.

text = "python is great python is fun python is powerful"

# 요구사항:
# 1. 문자열을 공백으로 분리하여 단어 리스트 만들기
# 2. 딕셔너리를 사용하여 각 단어의 등장 횟수 세기
# 3. 결과를 출력

# 힌트: 
# - 딕셔너리에 key가 없을 때 get() 함수의 default 값 활용
# - 또는 'word' in dict_name 으로 확인


a = text.split() # 단어 리스트 spilt() 은 문자열을 쪼개는 함수
b = {} #단어를 key, 등장횟수를 value 값으로 등장시키기 위해 빈 딕셔너리 생성

# b[x]는 b 딕셔너리에서 x의 미지수를 가지는 key의 value값을 출력한다는 의미
# 여기에 =으로 get을 사용해 key값으로 value값을 얻는 함수를 만들고
#(x,0)는 x일 때의 value 값을 얻고 만약 없으면 0을 출력한다는 의미
# +1은 현재 인식한 x값이 리스트에 있으면 1을 더하라는 의미
# get을 쓰는 이유는 처음에 key가 없기에 count를 쓰면 에러가 난다
for x in a:
    b[x] = b.get(x, 0) + 1

#f는 f-string(포맷 문자열)이라고 부른다
#쉽게 말하면 문자열 안에서 변수 값을 바로 집어넣게 해주는 표시
#즉, 괄호를 없애고 변수 값만 넣게 해주는 표시이다.
#이때 f가 괄호르 없애는 게 아니고 *변수 값만 꺼내서 보여주는 역할*
for a, c in b.items():
    print(f"{a}: {c}번")
```

```python
### 문제 4-1. 중첩 딕셔너리
#다음과 같은 중첩 딕셔너리를 다루는 문제를 해결하시오.


school = {
    '1반': {
        '김철수': {'국어': 85, '영어': 90, '수학': 78},
        '이영희': {'국어': 92, '영어': 88, '수학': 95}
    },
    '2반': {
        '박민수': {'국어': 88, '영어': 85, '수학': 90},
        '최지영': {'국어': 95, '영어': 92, '수학': 88}
    }
}

# 요구사항:
# 1. '1반' 김철수의 수학 점수를 출력하시오
# 2. '2반' 박민수의 평균 점수를 계산하여 출력하시오
# 3. 전교에서 국어 점수가 가장 높은 학생을 찾으시오
# 4. 각 반의 평균 점수를 계산하시오

school["1반"]["김철수"]["수학"]
print(school["1반"]["김철수"]["수학"])


#values는 value를 list로 만든다.
a = school['2반']["박민수"].values()
민수avg = sum(a) / len(a)
print(민수avg)


# 3. 전교 국어 점수 top
#반복문 전에는 변수가 존재해야 해서 변수를 먼저 설정해준다
top_student = ''
top_score = 0

#ban이라는 변수에 school의 value값을 리스트로 만들어 넣는걸 반복(for)
#ban 변수의 key, value를 각각 name과 scores 변수에 넣는걸 반복(for)
#top score를 0으로 설정하고 하나하나 반복하며 확인했을 때, 이전의 
#스코어 보다 높으면 새로운 top score로 갱신한다는 if 가정문 
#그렇게 top score가 지정되면 name의 변수에 저장된 key 값이 같이 저장되도록
#top_student = name 으로 식 설정
for ban in school.values():
    for name, scores in ban.items():
        if scores['국어'] > top_score:
            top_score = scores['국어']
            top_student = name

print(top_student, top_score)

# 4. 각 반 평균
#스쿨의 key[1반,2반] value 값을 클래스, 반 으로 나눠서 넣어준다
#반복문을 사용할 것이기에 변수값을 초기화 한다
#우리가 필요한 것은 반 딕셔너리의 점수이니 반value 값을 다시 리스트로 변환해 변수 스코어에 넣어준다
#스코어의 값에서 오로지 숫자인 점수 값만 필요하니 다시 리스트로 만들고 그것을 바로 sum으로 total 값을 구한다
#나누는 총 개수는 그냥 scores value의 개수로 해도 상관없으니 len으로 구해준다
# +=는 누적해서 더해주는 의미가 있으니 +=를 사용해 모든 value값을 더해주고 출력한다
for class_name, ban in school.items():
    total = 0
    count = 0
    for scores in ban.values():
        total += sum(scores.values())
        count += len(scores)
    print(class_name, total / count)
```

```python
### 문제 6. 두 반의 학생 명단 비교

#A반과 B반 학생 명단이 있습니다.


a_class = {'김철수', '이영희', '박민수', '최지영', '홍길동'}
b_class = {'박민수', '최지영', '정수민', '강호동', '이영희'}

#다음 질문에 답하시오:


# 1. 두 반 모두에 있는 학생 (교집합)

print("교집합 : " , a_class & b_class )

# 2. 두 반을 합친 전체 학생 (합집합)

print("합집합 : ", a_class | b_class)

# 3. A반에만 있는 학생 (차집합)

print( "A차집합 : ", a_class - b_class)

# 4. B반에만 있는 학생 (차집합)

print( "B차집합 : ", b_class - a_class)
```

```python
### 보너스 문제(선택). 영어-한글 사전 만들기
#간단한 영한사전 프로그램을 만드시오.

# 최소 10개 이상의 단어를 포함한 영한사전 딕셔너리 생성
# 사용자가 영어 단어를 입력하면 한글 뜻을 출력
# 없는 단어면 "사전에 없는 단어입니다" 출력
# 'quit' 입력 시 종료

dictionary = {
    'apple': '사과',
    'banana': '바나나',
    'orange': '오렌지',
    'computer': '컴퓨터',
    'python': '파이썬',
    'book': '책',
    'school': '학교',
    'student': '학생',
    'teacher': '선생님',
    'love': '사랑'
}
#while은 무한 반복문이다
#즉 특정 종료 프로그램을 넣지 않으면 계속 사용가능

##input()는 ()안의 문자열 메시지를 출력해서 보여주고 사용자가 입력 하면
#지정된 변수에 사용자가 입력한 변수를 저장하는 함수

## word가 'quit'이라는 문자열이 나오면 "프로그램 종료"라는 문자가 
# 출력되면서 종료된다는 의미
# 여기서 break는 무한 반복문인 while을 종료한다는 의미

## elif는 if 가정문의 부정 상황인 경우
# 여기서는 word가 'quit'가 아닌 경우

## 먼저 quit가 아닌 경우 input에 입력한 값이 dictionary 안에 있는지 확인
# 확인 후 만약 dictionary 안에 있다면 value 값을 출력

## else : 위의 모든 조건이 아닌 경우를 의미
# 위의 모든 조건이 아닌 경우 "사전에 없는 단어입니다"라는 문자열이 나오도록 출력 
while True:
    word = input("영어 단어 입력 (quit 입력 시 종료): ")

    if word == 'quit':
        print("프로그램 종료")
        break
    elif word in dictionary:
        print(f"{word} : {dictionary[word]}") # 반복문을 할 때 결과를 출력 당시 함수를 {}중괄호로 해줘야한다
    else:
        print("사전에 없는 단어입니다")
```
