```python
# 리스트 정렬 - sort / 오름차순으로 리스트의 수를 정리하게 해준다 / 문자의 오름차순 내리차순도 가능하다
a = [1, 4, 3, 2]
a.sort()
a
```

```python
# 인덱스 반환 - index / 인덱스 값을 나오게 하는 함수이다, 여기서 인덱스란 우리가 리스트의 순서를 0 1 2 3 이렇게 순서를 매기게 되는데 이때의 순서를 인덱스라고 한다
a = [1, 2, 3]
a.index(3)
```

```python
# 리스트에 요소 삽입 / insert(몇번째 인덱스, 삽입할 수)
a = [1, 2, 3]
a.insert(0, 4)
a

```

```python
# 리스트 요소 제거 -remove(제거할 값) 값이 중복된다면첫번째 나온 값만 제거되고 그 뒤의 값도 제거하고 싶다면 한번더 반복해서 실행하면 제거된다.
#but 현재 시트에서 a 의 값을 계속해서 선언하고 있기 때문에 현재 시트에서는 3이 첫번째만 제거되고 그대로 유지되어있다.
a = [1,2,3,1,2,3]
a.remove(3)
a
```

```python
# 리스트 요소 꺼내기 - pop 기본값이 -1로 설정되어 있어 현재는 a.pop()을 하면 마지막 값인 3이 출력 되고 현재 3값이 튀어나갔기 때문에 a를 출력하면 1,2만 출력된다
a = [1,2,3]
print(a.pop())
print(a)
```

```python
a = [1,2,3]
a.pop(1)
a
```

```python
# 리스트에 포함된 요소 x의 개수 세기 - count / 포함된 요소의 개수
a = [1, 2, 3, 1]
a.count(2)
```
