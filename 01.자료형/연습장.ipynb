````python
# ### 문제 7: 소수 판별기 
# 사용자가 입력한 숫자가 소수인지 판별하고, 1부터 해당 숫자까지의 모든 소수를 출력하는 프로그램을 작성하세요.

# **요구사항**:
# - for 문과 if 문 사용
# - 2 이상의 숫자만 입력 가능
# - 소수 판별 로직 구현

# **실행 예시**:
# ```
# 숫자를 입력하세요: 20
# 20은(는) 소수가 아닙니다.

# 1부터 20까지의 소수:
# 2, 3, 5, 7, 11, 13, 17, 19
# 총 8개의 소수가 있습니다.
# ```

# ---

N = int(input("소수를 입력하세요 : "))

while N < 2:
    N = int(input("2 이상의 숫자를 입력하시오."))

# 소수 판별
is_prime = True 
for i in range(2, N): #range(2,20) => [2,,.....,19]
    if N % i == 0:
        is_prime = False
        break


if is_prime == True:
    print(f"숫자를 입력하세요 : {N}")
    print(f"{N}은 소수입니다.")
else:
    print(f"{N}은 소수가 아닙니다.")
    print(f"숫자를 입력하세요 : {N}")

prime_list = []

for num in range(2, N+1):
    is_prime_num = True
    for i in range(2, num):
        if num % i == 0:
            is_prime_num = False
            break
    if is_prime_num:
        prime_list.append(num)

print(f"1부터 {N}까지의 소수:")
print(f"{prime_list}")
print(f"총 {len(prime_list)}개의 소수가 있습니다.")
````

````python
   # ### 문제 11: 종합 성적 관리 프로그램 
# 여러 학생의 성적을 관리하는 프로그램을 작성하세요.

# **요구사항**:
# - 최소 3명 이상의 학생 정보 입력
# - 각 학생별 3과목 점수 입력
# - 평균 계산 및 등급 판정
# - 전체 학생 중 최고점, 최저점 학생 출력
# - 과목별 평균 출력

# **실행 예시**:
# ```
# 학생 수를 입력하세요: 3

# [1번 학생]
# 이름: 김철수
# 국어: 85
# 영어: 90
# 수학: 88
# 평균: 87.67점 (등급: B)

# [2번 학생]
# 이름: 이영희
# 국어: 95
# 영어: 92
# 수학: 96
# 평균: 94.33점 (등급: A)

# [3번 학생]
# 이름: 박민수
# 국어: 78
# 영어: 82
# 수학: 75
# 평균: 78.33점 (등급: C)

# === 전체 통계 ===
# 최고 점수: 이영희 (94.33점)
# 최저 점수: 박민수 (78.33점)
# 국어 평균: 86.00점
# 영어 평균: 88.00점
# 수학 평균: 86.33점
# ```

# ---

# 학생 수
while True:
    N = int(input("학생 수 입력하시오."))
    if N < 3:
        print("다시 입력하시오")
    else:
        break

students=[]

for i in range(1, N+1):
    print(f"{i}번 학생")
    이름 = input("이름을 입력하시오.")
    국어 = int(input("점수를 입력하시오."))
    영어 = int(input("점수를 입력하시오."))
    수학 = int(input("점수를 입력하시오."))
    
    평균 = round(( 국어 + 영어 + 수학) / 3 ,2)


    if 평균 >= 90:
        등급 = "A"
    elif 평균 >= 80:
        등급 = "B"
    elif 평균 >= 70:
        등급 = "C"
    elif 평균 >= 60:
        등급 = "D"
    elif 평균 >= 50:
        등급 = "E"
    elif 평균 < 50:
        등급 = "F"


    students.append({
        "이름": 이름,
        "국어": 국어,
        "영어": 영어,
        "수학": 수학,
        "평균": 평균,
        "등급": 등급
    })
     
highest = max(students, key=lambda s: s["평균"])  
lowest = min(students, key=lambda s: s["평균"]) 

# MAX(딕셔너리, 키 : value) 의 값 중 max 값을 가져온다는 의미
#key=lambda 는 key의 값 중 임의의값이라는 의미로 그냥 변수임 , :는 key와 value 를 나누는 기준이고 , value의  값을 가져오는 방법은 딕셔너리 이름[키이름]

국어_avg = round(sum(s["국어"] for s in students) / N, 2)
영어_avg = round(sum(s["영어"] for s in students) / N, 2)
수학_avg = round(sum(s["수학"] for s in students) / N, 2)

print(f"학생 수를 입력하세요 : {N}")

print("\n=== 학생 성적 ===")
for s in students:
    print(f"이름: {s['이름']}")
    print(f"국어: {s['국어']}, 영어: {s['영어']}, 수학: {s['수학']}")
    print(f"평균: {s['평균']}점 (등급: {s['등급']})")
    print()

print("\n=== 전체 통계 ===")
print(f"최고 점수: {highest['이름']} ({highest['평균']}점)")
print(f"최저 점수: {lowest['이름']} ({lowest['평균']}점)")
print(f"국어 평균: {국어_avg}점")
print(f"영어 평균: {영어_avg}점")
print(f"수학 평균: {수학_avg}점")


````
